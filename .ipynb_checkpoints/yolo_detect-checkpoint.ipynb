```python
import cv2
import psycopg2
import time
import numpy as np
from ultralytics import YOLO
from datetime import datetime

# --- КОНФИГУРАЦИЯ ---
VIDEO_PATH = 'Модуль А/fallback2.mp4'
DB_URL = "dbname=user43 user=user43 password=m5q3x8tpc7vn host=109.123.153.227 port=5402"

# Твои наклонные линии (пиксели)
LINE_START_P1, LINE_START_P2 = (702, 457), (1236, 817)
LINE_END_P1, LINE_END_P2 = (991, 361), (1606, 474)

REAL_DIST_M = 15.0
LINE_THRESHOLD = 25  
PIXELS_PER_METER = 30
CRITICAL_DIST_M = 3.5
INCIDENT_COOLDOWN = 2

# --- ИНИЦИАЛИЗАЦИЯ ---
model = YOLO('yolo12s.pt')
cap = cv2.VideoCapture(VIDEO_PATH)

# --- БЕЗОПАСНОЕ ПОДКЛЮЧЕНИЕ К БД ---
db_active = False
try:
    # Установили жесткий таймаут 2 секунды, чтобы не ждать вечно
    conn = psycopg2.connect(DB_URL, connect_timeout=2)
    conn.autocommit = True
    cursor = conn.cursor()
    db_active = True
    print("✅ БАЗА ДАННЫХ: Подключено.")
except Exception:
    print("⚠️ БАЗА ДАННЫХ: Таймаут. Работаем в АВТОНОМНОМ режиме (логи в local_log.txt)")

entry_times = {}
processed_ids = set()
last_incident_time = {}

def point_line_distance(px, py, x1, y1, x2, y2):
    return abs((y2 - y1)*px - (x2 - x1)*py + x2*y1 - y2*x1) / np.sqrt((y2 - y1)**2 + (x2 - x1)**2)

def log_event(message):
    """Дублирует данные в консоль и текстовый файл"""
    print(message)
    with open("local_log.txt", "a", encoding="utf-8") as f:
        f.write(f"{datetime.now()} | {message}\n")

log_event("🚀 Мониторинг запущен.")

try:
    while cap.isOpened():
        success, frame = cap.read()
        if not success: break

        now_ts, now_dt = time.time(), datetime.now()
        results = model.track(frame, persist=True, verbose=False, classes=[2, 3, 5, 7])
        annotated_frame = results[0].plot()

        if results[0].boxes.id is not None:
            boxes = results[0].boxes.xyxy.cpu().numpy().astype(float)
            ids = results[0].boxes.id.cpu().numpy().astype(int)
            clss = results[0].boxes.cls.cpu().numpy().astype(int)
            active_objects = []

            for i in range(len(ids)):
                x1, y1, x2, y2 = boxes[i]
                t_id, cx, cy = int(ids[i]), (x1 + x2) / 2, (y1 + y2) / 2
                active_objects.append({'id': t_id, 'cx': cx, 'cy': cy})

                d_start = point_line_distance(cx, cy, *LINE_START_P1, *LINE_START_P2)
                d_end = point_line_distance(cx, cy, *LINE_END_P1, *LINE_END_P2)

                # Вход на замер
                if d_start < LINE_THRESHOLD and t_id not in entry_times:
                    entry_times[t_id] = now_ts

                # Выход и расчет скорости
                if d_end < LINE_THRESHOLD and t_id in entry_times and t_id not in processed_ids:
                    duration = now_ts - entry_times[t_id]
                    if duration > 0.3:
                        speed = (REAL_DIST_M / duration) * 3.6
                        processed_ids.add(t_id)
                        
                        msg = f"SPEED: ID {t_id} | Тип: {model.names[clss[i]]} | {speed:.1f} км/ч"
                        log_event(msg)
                        
                        if db_active:
                            try:
                                cursor.execute("INSERT INTO full_tracking_data (video_id, track_id, speed_km_h) VALUES (12, %s, %s)", (t_id, float(speed)))
                            except: db_active = False

            # Проверка опасных сближений
            for i in range(len(active_objects)):
                for j in range(i + 1, len(active_objects)):
                    o1, o2 = active_objects[i], active_objects[j]
                    dist_m = (((o1['cx']-o2['cx'])**2 + (o1['cy']-o2['cy'])**2)**0.5) / PIXELS_PER_METER
                    
                    if dist_m < CRITICAL_DIST_M:
                        pair = tuple(sorted((o1['id'], o2['id'])))
                        if pair not in last_incident_time or (now_ts - last_incident_time[pair]) > INCIDENT_COOLDOWN:
                            log_event(f"⚠️ DANGER: Сближение {o1['id']} и {o2['id']} ({dist_m:.2f}м)")
                            if db_active:
                                try:
                                    cursor.execute("INSERT INTO dangerous_incidents (track_id1, track_id2, distance) VALUES (%s, %s, %s)", (o1['id'], o2['id'], float(dist_m)))
                                except: db_active = False
                            last_incident_time[pair] = now_ts

        # Визуализация
        cv2.line(annotated_frame, LINE_START_P1, LINE_START_P2, (0, 255, 0), 2) # Зеленая - СТАРТ
        cv2.line(annotated_frame, LINE_END_P1, LINE_END_P2, (0, 0, 255), 2)   # Красная - ФИНИШ
        
        cv2.imshow('Traffic Control', cv2.resize(annotated_frame, (960, 540)))
        if cv2.waitKey(1) & 0xFF == ord('q'): break

finally:
    cap.release()
    cv2.destroyAllWindows()
    if db_active:
        cursor.close()
        conn.close()
```
