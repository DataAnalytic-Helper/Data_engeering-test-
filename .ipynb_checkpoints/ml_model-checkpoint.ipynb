```python
import pandas as pd
import numpy as np
import psycopg2
import time
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from datetime import datetime, timedelta

# --- КОНФИГУРАЦИЯ ---
DB_URL = "dbname=user43 user=user43 password=m5q3x8tpc7vn host=2.nntc.nnov.ru port=5402"
CSV_PATH = 'Модуль В/traffic_full_30min.csv'
HORIZON = 60 # Фиксированный горизонт по КЗ (60 минут)
```

```python
try:
    df = pd.read_csv(CSV_PATH)
    # Приводим к минутной интенсивности
    df['moment_cars'] = df['intensity_30min'] / 30.0 
    # Сдвиг таргета на 60 минут (предполагаем, что шаг в CSV - 30 минут, значит сдвиг на 2 строки)
    df['target'] = df['moment_cars'].shift(-2) 
    
    valid_data = df.dropna()
    X = valid_data[['moment_cars']]
    y = valid_data['target']
    
    # Обучение модели
    model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=4, random_state=42)
    model.fit(X, y)
    
    # Оценка точности
    preds = model.predict(X)
    mae = mean_absolute_error(y, preds)
    rmse = mean_squared_error(y, preds) ** 0.5
    r2 = r2_score(y, preds)
    
    print("-" * 40)
    print(f"✅ Модель обучена. Горизонт: +{HORIZON} мин")
    print("МЕТРИКИ КАЧЕСТВА:")
    print("MAE:  {0:7.2f}".format(mae))
    print("RMSE: {0:7.2f}".format(rmse))
    print("R2:   {0:7.2f}".format(r2))
    print("-" * 40)
    
except Exception as e:
    print(f"❌ Ошибка при обучении: {e}")
    exit()
```

```python
# 2. Прогнозирование
def predict_and_store():
    try:
        with psycopg2.connect(DB_URL) as conn:
            with conn.cursor() as cur:
                # Берем реальный факт за последние 5 минут для сглаживания
                cur.execute("""
                    SELECT COUNT(DISTINCT track_id), COALESCE(AVG(speed_km_h), 0) 
                    FROM full_tracking_data 
                    WHERE detection_time >= NOW() - INTERVAL '5 minute'
                """)
                row = cur.fetchone()
                
                # Защита от пустых ответов
                curr_count = float(row[0]) if row[0] is not None else 0.0
                curr_speed = float(row[1]) if row[1] is not None else 0.0
                
                now = datetime.now()
                target_time = now + timedelta(minutes=HORIZON)
                
                # Создаем DataFrame для предикта, чтобы не было Warning'ов
                input_df = pd.DataFrame([[curr_count]], columns=['moment_cars'])
                pred_val = float(model.predict(input_df)[0])
                pred_val = max(0.0, pred_val) # Интенсивность не может быть отрицательной
                
                # Вставка в БД
                cur.execute("""
                    INSERT INTO traffic_predictions 
                    (target_time, horizon_minutes, predicted_intensity, predicted_avg_speed) 
                    VALUES (%s, %s, %s, %s)
                """, (target_time, HORIZON, pred_val, curr_speed))
                
                print(f"[{now.strftime('%H:%M:%S')}] 🔮 Прогноз записан | Факт сейчас: {int(curr_count)} ТС | Прогноз через {HORIZON}м: {pred_val:.1f} ТС")
                
    except Exception as e:
        print(f"❌ Ошибка БД в цикле прогноза: {e}")

# Бесконечный цикл с обработкой прерывания
try:
    while True:
        predict_and_store()
        time.sleep(30) # Запуск по расписанию (раз в 30 секунд для демо)
except KeyboardInterrupt:
    print("\n🛑 Цикл прогнозирования остановлен пользователем.")
```

```python
import cv2
import psycopg2
import time
import numpy as np
import os
import requests
import boto3
from requests.auth import HTTPBasicAuth
from ultralytics import YOLO
from datetime import datetime

# ==========================================
# 1. КОНФИГУРАЦИЯ СИСТЕМЫ
# ==========================================
# Видеопоток
RTMP_URL = 'rtmp://nntc.nnov.ru:5566/stream'

# База данных
DB_URL = "dbname=user43 user=user43 password=m5q3x8tpc7vn host=2.nntc.nnov.ru port=5402"

# Облачное хранилище (MinIO/S3)
S3_ENDPOINT = "http://2.nntc.nnov.ru:9000"
S3_ACCESS_KEY = "ВАШ_КЛЮЧ_MINIO"     # ЗАМЕНИТЬ на ключи от MinIO
S3_SECRET_KEY = "ВАШ_СЕКРЕТ_MINIO"   # ЗАМЕНИТЬ на ключи от MinIO
S3_BUCKET_NAME = "traffic-frames"
UPLOAD_INTERVAL_SEC = 30 * 60  # 30 минут

# Airflow Оркестрация
AIRFLOW_API_URL = "http://localhost:8080/api/v1/dags/traffic_monitoring_pipeline/dagRuns"
AIRFLOW_AUTH = HTTPBasicAuth('admin', 'admin')
AIRFLOW_TRIGGER_INTERVAL = 5 * 60  # Триггер ML-прогноза каждые 5 минут

# Калибровка (Наклонные линии для расчета скорости)
# Указываем координаты точек (x, y) для отрисовки линий на дороге
LINE_1_START, LINE_1_END = (250, 350), (1050, 350) # Верхняя линия
LINE_2_START, LINE_2_END = (100, 550), (1200, 550) # Нижняя линия (шире из-за перспективы)

LINE_1_Y = 350 
LINE_2_Y = 550  
REAL_DIST_M = 15.0 # Реальное расстояние в метрах
PIXELS_PER_METER = abs(LINE_2_Y - LINE_1_Y) / REAL_DIST_M

# Правила безопасности
CRITICAL_DIST_M = 3.5
INCIDENT_COOLDOWN = 2

# ==========================================
# 2. ИНИЦИАЛИЗАЦИЯ
# ==========================================
model = YOLO('yolo12s.pt')
cap = cv2.VideoCapture(RTMP_URL)

# Настройка S3 клиента
s3_client = boto3.client('s3', 
                         endpoint_url=S3_ENDPOINT,
                         aws_access_key_id=S3_ACCESS_KEY,
                         aws_secret_access_key=S3_SECRET_KEY)

conn = psycopg2.connect(DB_URL)
conn.autocommit = True
cursor = conn.cursor()

# Переменные состояния
entry_times = {} 
processed_ids = set()
last_incident_time = {}
object_speeds = {} # Храним скорости для отрисовки на кадре: {id: speed}

last_s3_upload = time.time()
last_airflow_trigger = time.time()

print("🚀 Система запущена. Чтение RTMP-потока...")

# ==========================================
# 3. ОСНОВНОЙ ЦИКЛ ОБРАБОТКИ
# ==========================================
try:
    while cap.isOpened():
        success, frame = cap.read()
        if not success: 
            print("⚠️ Потерян кадр из потока. Переподключение...")
            time.sleep(1)
            cap = cv2.VideoCapture(RTMP_URL) # Попытка переподключения к RTMP
            continue
        
        now_ts = time.time()
        now_dt = datetime.now()
        
        # Детекция (только транспорт: 2-car, 3-motorcycle, 5-bus, 7-truck)
        results = model.track(frame, persist=True, verbose=False, classes=[2, 3, 5, 7])
        
        # Получаем базовую картинку с рамками
        annotated_frame = results[0].plot()

        # Отрисовка наклонных калибровочных линий (как на фото)
        cv2.line(annotated_frame, LINE_1_START, LINE_1_END, (0, 255, 255), 3) # Желтая
        cv2.line(annotated_frame, LINE_2_START, LINE_2_END, (0, 255, 255), 3) # Желтая
        cv2.putText(annotated_frame, "15 METERS", (LINE_1_START[0], LINE_1_START[1] - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)

        active_objects = []

        if results[0].boxes.id is not None:
            boxes = results[0].boxes.xyxy.cpu().numpy().astype(float)
            ids = results[0].boxes.id.cpu().numpy().astype(int)
            clss = results[0].boxes.cls.cpu().numpy().astype(int)
            
            for i in range(len(ids)):
                x1, y1, x2, y2 = boxes[i]
                t_id = int(ids[i])
                cx, cy = (x1 + x2) / 2, (y1 + y2) / 2
                obj_name = model.names[clss[i]]
                active_objects.append({'id': t_id, 'cx': cx, 'cy': cy, 'name': obj_name})

                # ЛОГИКА СКОРОСТИ
                # Машина пересекла верхнюю линию
                if abs(cy - LINE_1_Y) < 30 and t_id not in entry_times:
                    entry_times[t_id] = now_ts
                
                # Машина пересекла нижнюю линию
                if cy > LINE_2_Y and t_id in entry_times and t_id not in processed_ids:
                    duration = now_ts - entry_times[t_id]
                    if duration > 0:
                        speed = (REAL_DIST_M / duration) * 3.6 # km/h
                        object_speeds[t_id] = round(speed, 1) # Сохраняем для отрисовки
                        
                        # Запись в БД
                        cursor.execute("""
                            INSERT INTO full_tracking_data 
                            (video_id, track_id, object_type, detection_time, x_cord_m, y_cord_m, speed_km_h) 
                            VALUES (1, %s, %s, %s, %s, %s, %s)
                        """, (t_id, obj_name, now_dt, float(cx), float(cy), float(speed)))
                        processed_ids.add(t_id)

                # ОТРИСОВКА СКОРОСТИ НАД МАШИНОЙ
                if t_id in object_speeds:
                    cv2.putText(annotated_frame, f"{object_speeds[t_id]} km/h", 
                                (int(x1), int(y1) - 10), 
                                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

            # ТРИГГЕР СБЛИЖЕНИЙ
            for i in range(len(active_objects)):
                for j in range(i + 1, len(active_objects)):
                    obj1, obj2 = active_objects[i], active_objects[j]
                    dist_px = ((obj1['cx'] - obj2['cx'])**2 + (obj1['cy'] - obj2['cy'])**2)**0.5
                    dist_m = round(dist_px / PIXELS_PER_METER, 2)
                    
                    if dist_m < CRITICAL_DIST_M:
                        pair = tuple(sorted((obj1['id'], obj2['id'])))
                        if pair not in last_incident_time or (now_ts - last_incident_time[pair]) > INCIDENT_COOLDOWN:
                            cursor.execute("""
                                INSERT INTO dangerous_incidents (time, track_id1, track_id2, distance) 
                                VALUES (%s, %s, %s, %s)
                            """, (now_dt, obj1['id'], obj2['id'], float(dist_m)))
                            last_incident_time[pair] = now_ts

                            # Отрисовка красной линии между нарушителями
                            cv2.line(annotated_frame, 
                                     (int(obj1['cx']), int(obj1['cy'])), 
                                     (int(obj2['cx']), int(obj2['cy'])), 
                                     (0, 0, 255), 2)

        # ==========================================
        # 4. ИНТЕГРАЦИИ (Airflow и S3)
        # ==========================================
        
        # А) Триггер Airflow (Event-Driven ML) каждые 5 минут
        if now_ts - last_airflow_trigger >= AIRFLOW_TRIGGER_INTERVAL:
            try:
                requests.post(AIRFLOW_API_URL, auth=AIRFLOW_AUTH, json={"conf": {}}, timeout=2)
                print(f"[{now_dt.strftime('%H:%M:%S')}] 🚀 Сигнал в Airflow отправлен.")
            except Exception as e:
                print(f"⚠️ Ошибка API Airflow: {e}")
            last_airflow_trigger = now_ts

        # Б) Загрузка кадра в MinIO/S3 каждые 30 минут
        if now_ts - last_s3_upload >= UPLOAD_INTERVAL_SEC:
            fname = f"frame_{now_dt.strftime('%Y%m%d_%H%M%S')}.jpg"
            cv2.imwrite(fname, annotated_frame)
            try:
                # Отправка файла в MinIO бакет
                s3_client.upload_file(fname, S3_BUCKET_NAME, f"dashboard_screens/{fname}")
                print(f"[{now_dt.strftime('%H:%M:%S')}] ☁️ Кадр успешно загружен в S3: {S3_ENDPOINT}/{S3_BUCKET_NAME}/dashboard_screens/{fname}")
            except Exception as e:
                print(f"⚠️ Ошибка загрузки в S3: {e}")
            finally:
                if os.path.exists(fname): os.remove(fname) # Удаляем с локального диска
            last_s3_upload = now_ts

        # Вывод видео (для отладки можно закомментировать на сервере)
        cv2.imshow('Traffic Monitor (RTMP)', cv2.resize(annotated_frame, (960, 540)))
        if cv2.waitKey(1) & 0xFF == ord('q'): 
            break

finally:
    cap.release()
    cv2.destroyAllWindows()
    cursor.close()
    conn.close()
    print("🛑 Система остановлена.")
```
