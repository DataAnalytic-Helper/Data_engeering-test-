````python
# МОДУЛЬ А: Проектирование архитектуры инфраструктуры данных

**Конкурсант:** Анастасия  
**Дата:** 01.04.2026  
**Проект:** Интеллектуальная платформа мониторинга транспортных потоков

---

## 1. Анализ требований

### 1.1 Функциональные требования

1. **Потоковая обработка видео** - система должна захватывать видеопоток с камер в реальном времени и выполнять детекцию транспортных средств с помощью нейросети.
2. **Классификация объектов** - распознавание типов транспортных средств (легковые, грузовые, автобусы, мотоциклы, велосипеды, пешеходы) с фиксацией параметров:
   - координаты bounding box (`x1, y1, x2, y2`);
   - временная метка;
   - скорость движения (рассчитывается на основе трекинга).
3. **Накопление исторических данных** - система должна обеспечивать хранение полной истории детекций не менее 180 дней для построения прогнозных моделей.
4. **Формирование аналитических витрин** - подготовка данных для ответа на ключевые бизнес-вопросы: интенсивность, структура потока, средняя скорость, прогноз загруженности.

### 1.2 Нефункциональные требования

1. **Целостность данных** - исключение некорректных значений (NULL, выбросы) в критических полях (скорость, тип ТС) на этапах загрузки и трансформации.
2. **Масштабируемость** - возможность подключения новых камер и источников данных без изменения архитектуры и перестройки существующих пайплайнов.
3. **Актуальность данных** - обновление аналитических витрин в режиме *Near Real-Time* (задержка не более 2 минут).
4. **Производительность** - обработка видеопотока без сбоев при нагрузке >=30 кадров в секунду на одну камеру.
5. **Надёжность** - устойчивая работа при длительной непрерывной обработке потоковых данных (автоматическое восстановление после сбоев).

---

## 2. Выбор архитектуры и технологий

### 2.1 Используемый стек технологий

| Компонент            | Выбранная технология |
|----------------------|----------------------|
| Хранилище данных     | MySQL (DWH)          |
| Потоковая обработка  | Python (OpenCV, YOLO)|
| Пакетная обработка   | Python (pandas)      |
| Оркестрация          | Apache Airflow       |
| Визуализация         | Yandex DataLens / BI |

### 2.2 Обоснование выбора

- **MySQL** - централизованное хранилище с поддержкой ACID, подходит для роли DWH в гибридной архитектуре. Обеспечивает единую версию правды и простую интеграцию с BI-инструментами.
- **YOLO (You Only Look Once)** - современная нейросетевая модель детекции объектов. Высокая скорость (>=30 FPS) и точность (mAP ~50%) позволяют обрабатывать видеопоток в реальном времени. Встроенный трекинг упрощает подсчёт уникальных ТС и расчёт скорости.
- **Python + OpenCV** - гибкая экосистема для захвата видео, интеграции с YOLO и расчёта метрик. Разделение на потоковую и пакетную обработку (pandas) позволяет удовлетворить требования *real-time* и глубокого исторического анализа.
- **Apache Airflow** - стандарт оркестрации: позволяет автоматизировать пакетные ETL-процессы (загрузка, очистка, обогащение, расчёт витрин) по расписанию, обеспечивая воспроизводимость и мониторинг.
- **Yandex DataLens** - прямое требование конкурсного задания. Бесплатный отечественный инструмент, поддерживающий подключение к MySQL и работу с интерактивными дашбордами в реальном времени.

---

## 3. Проектирование архитектуры данных

### 3.1 Архитектурный подход

Применена **гибридная архитектура**, сочетающая:

- **Data Lake** - для хранения сырых видеоданных и неструктурированных файлов (например, кадров с детекциями).
- **Data Warehouse (DWH)** - на основе модели "звезда" для структурированной аналитики.

Это позволяет сохранять историческую точность (Data Lake) и обеспечивать высокую производительность аналитических запросов (DWH).

### 3.2 Слои данных

| Слой                | Назначение                                   | Таблицы / хранилище          |
|---------------------|----------------------------------------------|------------------------------|
| **RAW (сырой)**     | Хранение исходных детекций и видеокадров     | `detections_raw`, файловая система (Data Lake) |
| **CORE (ядро)**     | Нормализованные данные: камеры, кадры, ТС    | `cameras`, `frames`, `vehicles` |
| **DATA MART (витрины)** | Агрегированные метрики для BI              | `traffic_metrics`, `vehicle_stats`, `congestion` |

### 3.3 Потоки данных

1. **Потоковый контур**  
   `Видеопоток` -> `YOLO (детекция + трекинг)` -> `Потоковый обработчик (Python)` -> **RAW-слой** -> **CORE-слой** -> **Витрины** -> `Дашборд (DataLens)`

2. **Пакетный контур (Airflow)**  
   `RAW-слой` -> `Очистка / обогащение` -> `CORE-слой` -> `Расчёт витрин` -> `Дашборд (исторические данные)`

---

## 4. Моделирование данных

### 4.1 Концептуальная модель

**Сущности и связи:**
- **Камера** (camera_id, location)  
- **Кадр** (frame_id, camera_id, timestamp)  
- **Детекция** (detection_id, frame_id, vehicle_id, координаты, скорость, confidence)  
- **Транспортное средство** (vehicle_id, vehicle_type)

Связи:  
`Camera` 1 : N `Frame`  
`Frame` 1 : N `Detection`  
`Detection` N : 1 `Vehicle`

### 4.2 Логическая модель

**Атрибуты сущностей:**

| Сущность      | Атрибуты |
|---------------|----------|
| Camera        | camera_id (PK), location, road_name |
| Frame         | frame_id (PK), camera_id (FK), timestamp |
| Vehicle       | vehicle_id (PK), vehicle_type (car, truck, bus, motorcycle, bicycle, person) |
| Detection     | detection_id (PK), frame_id (FK), vehicle_id (FK), x1, y1, x2, y2, confidence, speed |

### 4.3 Физическая модель (SQL DDL)

Код для развёртывания в MySQL (MyAdmin):

```sql
-- Создание базы данных
CREATE DATABASE IF NOT EXISTS traffic_dwh;
USE traffic_dwh;

-- Справочник типов ТС
CREATE TABLE vehicle_type (
    type_id   TINYINT UNSIGNED AUTO_INCREMENT PRIMARY KEY,
    type_name VARCHAR(20) NOT NULL UNIQUE
);

-- Таблица камер
CREATE TABLE cameras (
    camera_id   INT AUTO_INCREMENT PRIMARY KEY,
    location    VARCHAR(255) NOT NULL,
    road_name   VARCHAR(255)
);

-- Таблица кадров
CREATE TABLE frames (
    frame_id    BIGINT AUTO_INCREMENT PRIMARY KEY,
    camera_id   INT NOT NULL,
    timestamp   DATETIME NOT NULL,
    FOREIGN KEY (camera_id) REFERENCES cameras(camera_id)
);

-- Таблица транспортных средств (треков)
CREATE TABLE vehicles (
    vehicle_id  BIGINT AUTO_INCREMENT PRIMARY KEY,
    type_id     TINYINT UNSIGNED NOT NULL,
    FOREIGN KEY (type_id) REFERENCES vehicle_type(type_id)
);

-- Таблица детекций (RAW слой)
CREATE TABLE detections (
    detection_id BIGINT AUTO_INCREMENT PRIMARY KEY,
    frame_id     BIGINT NOT NULL,
    vehicle_id   BIGINT NOT NULL,
    x1           FLOAT NOT NULL,
    y1           FLOAT NOT NULL,
    x2           FLOAT NOT NULL,
    y2           FLOAT NOT NULL,
    confidence   FLOAT NOT NULL CHECK (confidence BETWEEN 0 AND 1),
    speed        FLOAT COMMENT 'km/h, may be NULL initially',
    load_time    TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    FOREIGN KEY (frame_id) REFERENCES frames(frame_id),
    FOREIGN KEY (vehicle_id) REFERENCES vehicles(vehicle_id)
);

-- Индексы для производительности
CREATE INDEX idx_detections_frame ON detections(frame_id);
CREATE INDEX idx_frames_camera_time ON frames(camera_id, timestamp);
````
